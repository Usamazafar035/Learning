```python
# Calculate True posative rate and threshhold
fpr, tpr, threshold = roc_curve(n_test, n_prob_posative )
```

```python
fpr
```

```python
#create a function for ploting using matplotlib
import matplotlib.pyplot as plt
def plot_roc(fpr,tpr):
    #plot roc curve
    plt.plot(fpr, tpr, color = 'red', label= "AUC")
    #customiztion
    plt.title('Area ubder the ROC curve')
    plt.xlabel('False Posative Rate(fpr)')
    plt.ylabel('True Posative Rate(tpr)')
    plt.legend()
    plt.show()

plot_roc(fpr, tpr)

```

```python
def evalute_hyper_matrics(p_true, p_predicted):
    '''this function returns the evalation matrics'''
    accuracy = accuracy_score(p_true,p_predicted)
    precision = precision_score(p_true, p_predicted)
    recall = recall_score(p_true, p_predicted)
    f1 = f1_score(p_true, p_predicted)
    
    metrics_dict ={"Accuracy": accuracy,
             'precision': precision,
              'recall' :  recall,     
             'f1': f1}

    print(f"accuracy_Score : {round(accuracy,2)}")
    print(f"precision : {round(precision,2)}")
    print(f"recall : {round(recall,2)}")
    print(f"f1 : {round(f1)}")
  
    return metrics_dict

```

```python
#Train
#Validate
#Test
Heart_DataN= pd.read_csv('heart.csv')
mixed_heart_data = Heart_DataN.sample(frac=1)
#split X,Y
o= mixed_heart_data.drop('target', axis=1)
p= mixed_heart_data['target']
#splitting data into 3 parts now: train 70%, Validate 15%, test 15%
train_split= round(0.7*len(mixed_heart_data))
valid_split=round( train_split + 0.15*len(mixed_heart_data))
o_train, p_train  = o[:train_split], p[:train_split]
o_valid, p_valid = o[train_split:valid_split], p[train_split:valid_split]
o_test, p_test = o[valid_split:], p[valid_split:]

```

```python
hparams = { "max_depth" : [None, 5,10,20,30],
           "min_samples_leaf" : [1,2,4],
           "min_samples_split" : [2,4,6],
           "n_estimators" : [10,100,200,500,1000,1200],
           "max_features": ['auto', 'sqrt']}
#split into o and p
o= mixed_heart_data.drop('target', axis=1)
p= mixed_heart_data['target']
o_train, o_test, p_train, p_test = train_test_split(o,p,test_size=0.20)

clf = RandomForestClassifier(n_jobs=1)
```

```python
from sklearn.model_selection import RandomizedSearchCV
rscv_clf =  RandomizedSearchCV(estimator = clf, 
                              param_distributions=hparams,
                              n_iter= 10, 
                              cv=5,
                              verbose=2)
```

```python
rscv_clf .fit(o_train, p_train)
```

```python
rscv_clf.best_params_
```

```python
#with RSCV
#Running prediction
rscv_o_predicted = rscv_clf.predict(o_test)
#Evalute the prediction
fourth_result = evalute_hyper_matrics(p_test, rscv_o_predicted)
```

```python
#Tunig Hyperparameters using Grid Search Cross Validstion(GSCV)
from sklearn.model_selection import GridSearchCV
hparams2 = { "max_depth" : [5,10,20],
           "min_samples_leaf" : [2,4],
           "min_samples_split" : [2,4],
           "n_estimators" : [100,200,500],
           "max_features": ['auto', 'sqrt']}
gscv_clf = GridSearchCV(estimator = clf, 
                              param_grid=hparams2, 
                              cv=5,
                              verbose=2)
gscv_clf.fit(o_train, p_train)
```
