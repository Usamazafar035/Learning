pyplot vs Object Oriented Method

```python
walmart = pd.read_csv('Walmart.csv')
walmart
```

```python
walmart.head(5)
```

```python
over_40 = walmart[walmart['Temperature']>40]
over_40
```

```python
over_40_round = over_40['Fuel_Price'].round().astype(int)
over_40_round 
```

```python
#pyplot method
over_40.plot( kind = 'scatter' , 
             x = 'Weekly_Sales',
             y= 'Fuel_Price',
             c ='Store');
```

```python
#OBJECT ORIENTED APPROCH
fig, ax = plt.subplots(figsize= (6,5))
over_40.plot(kind= 'scatter',
            x= 'Weekly_Sales', 
            y= 'Fuel_Price',
            c= 'Store', 
            ax=ax);
```

#entire life cycle of oo plot

```python
fig, ax = plt.subplots(figsize= (6,5))
scatter_plot =ax.scatter(x= walmart['Weekly_Sales'], 
            y= walmart['Fuel_Price'],
            c= walmart['Store']);
ax.set( title ='sales of store',
       xlabel = 'Weekly Sales',
       ylabel = 'Fuel Price')

ax.legend(*scatter_plot.legend_elements(), title = 'Store')
scatter_plot;
```

Life cycle of OO plot advanced

```python
#create plot area
fig, (plot1, plot2) = plt.subplots(nrows= 2, 
                                   ncols= 1, 
                                   figsize= (10,10),
                                 sharex=True ) #to common the same axis
#add data to plot1
scatter_plot = plot1.scatter(x= walmart['Weekly_Sales'], 
            y= walmart['Fuel_Price'],
            c= walmart['Store'],
            cmap= 'winter')
plot1.set(title = 'sales',
          #xlabel = 'Weekly_Sales' ,
          ylabel = 'Store')
plot1.legend(*scatter_plot.legend_elements (), title= 'sales');
#Temperature
scatter_plot = plot2.scatter( x=  walmart['Weekly_Sales'],
                             y = walmart['Temperature'],
                             c= walmart['Store'],
                            cmap= 'plasma') #used to change the colour 
plot2.set(title = 'sales2',
          xlabel = 'weekly sales',
          ylabel = 'Temperature')

plot2.legend(*scatter_plot.legend_elements(), title= 'sales2');
#customization
plot2.axhline( y = walmart['Temperature'].mean(), linestyle = "--");
plot1.axhline(y= walmart['Store'].mean(), linestyle = '--');
#colouring plots
plt.style.use('seaborn-v0_8-darkgrid')
#Naming tha figure
fig.suptitle('walmart Data', fontsize= 32, fontweight = 'bold' )
#save fig
fig.savefig('Walmart Data Report.png')
```
